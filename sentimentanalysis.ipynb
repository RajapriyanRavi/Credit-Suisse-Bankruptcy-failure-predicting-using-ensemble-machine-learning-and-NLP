```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
import pandas as pd

# Load the text file
def load_text(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

# Initialize the BERT tokenizer and model
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)

# Use pipeline for sentiment analysis
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

# Function to segment and analyze text
def analyze_text(text):
    # Break the text into manageable parts
    max_length = 512  # Model's maximum input length
    parts = [text[i: i + max_length] for i in range(0, len(text), max_length)]

    results = []
    for part in parts:
        # Ensure part has substantial content
        if len(part) > 50:  # Simple filter to avoid very short parts
            result = nlp(part)
            results.extend([(part, res['label'], res['score']) for res in result])
    return results

# Load and analyze the text
text = load_text('2007.txt')
sentiment_results = analyze_text(text)

# Convert results to DataFrame for better visualization and save it to a CSV file
df_results = pd.DataFrame(sentiment_results, columns=['Part', 'Sentiment', 'Confidence'])
df_results.to_csv('2007_sentiment_analysis.csv', index=False)

print("Sentiment analysis completed and saved to 2007_sentiment_analysis.csv.")
```

```python
import pandas as pd
import glob
import os

# Define the pattern for your CSV files
files = glob.glob("*.csv")  # Adjust the path if files are in a subdirectory

# Create an empty DataFrame to store all data
combined_data = pd.DataFrame()

# Loop through each file and append it to the combined DataFrame
for file in files:
    # Extract the year from the filename assuming it is the first part of the file
    filename = os.path.basename(file)  # Get only the filename
    year = filename.split('_')[0]  # Extract the year assuming format is 'YEAR_sentiment_analysis.csv'

    # Ensure the extracted part is a valid year
    try:
        year = int(year)  # Convert year to integer
    except ValueError:
        print(f"Skipping file {filename} as it doesn't contain a valid year.")
        continue  # Skip files without valid years

    # Read the CSV file and add the year column
    temp_df = pd.read_csv(file)  # Read CSV instead of Excel
    temp_df['Year'] = year  # Add a 'Year' column to the DataFrame
    combined_data = pd.concat([combined_data, temp_df], ignore_index=True)

# Save the combined data to a new CSV file
combined_data.to_csv("combined_sentiment_analysis.csv", index=False)
print("All files combined into 'combined_sentiment_analysis.csv'")
```

```python
# Ensure the Sentiment column is processed correctly
if data['Sentiment'].dtype == 'object':
    # Convert strings to numeric values
    data['Sentiment'] = data['Sentiment'].astype(str).str.extract(r'(\d)').astype(int)
else:
    # If already numeric, ensure it's in integer format
    data['Sentiment'] = data['Sentiment'].astype(int)

# Verify the conversion
print(data.dtypes)
print(data.head())
```
